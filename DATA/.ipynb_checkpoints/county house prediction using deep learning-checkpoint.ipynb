# KING COUNTY HOUSE PRICE PREDICTION USING ARTIFICIAL NEURAL NETWORK

In this project, I will be building a deep learning regression model that takes in some features of a house and predict the price of the house with high level of accuracy. This will be achieved using dataset of houses sold between May 2014 and May 2015 in King County, including Seatle. The dataset was downloaded from https://www.kaggle.com/harlfoxem/housesalesprediction. Brief description of the features are given below:

## FEATURE COLUMNS
    
* id - Unique ID for each home sold
* date - Date of the home sale
* price - Price of each home sold
* bedrooms - Number of bedrooms
* bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - Square footage of the apartments interior living space
* sqft_lot - Square footage of the land space
* floors - Number of floors
* waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not
* view - An index from 0 to 4 of how good the view of the property was
* condition - An index from 1 to 5 on the condition of the apartment,
* grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - The square footage of the interior housing space that is above ground level
* sqft_basement - The square footage of the interior housing space that is below ground level
* yr_built - The year the house was initially built
* yr_renovated - The year of the house’s last renovation
* zipcode - What zipcode area the house is in
* lat - Lattitude
* long - Longitude
* sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

### Data appear to be ready for model training

## TRAIN TEST SPLIT AND SCALING

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

## CREATING THE MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

### I shall set the callbacto early_stop to avoid overfitting

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=4)
```

## TRAINING THE MODEL

```python
# We train in batches because the dataset is large. Baches should be numbers in power of 2.
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400,
         callbacks=[early_stop])
```

## MODEL EVALUATION ON TEST DATA

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
X_test
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
# Our predictions
plt.scatter(y_test,predictions)
# Perfect predictions
plt.plot(y_test,y_test,'r')
```

```python
errors = y_test.values.reshape(6480, 1) - predictions
```

```python
sns.distplot(errors)
```

## CODE FOR DEPLOYMENT

```python
model = load_model("county_house_price_prediction_model.h5")
scaler = joblib.load("county_house_price_prediction_scaler.pkl")
```

```python
def return_prediction(model,scaler,sample_json):
    
    b = sample_json['bedrooms']
    c = sample_json['bathrooms']
    d = sample_json['sqft_living']
    e = sample_json['sqft_lot']
    f = sample_json['floors']
    g = sample_json['waterfront']
    h = sample_json['view']
    i = sample_json['condition']
    j = sample_json['grade']
    k = sample_json['sqft_above']
    l = sample_json['sqft_basement']
    m = sample_json['yr_built']
    n = sample_json['yr_renovated']
    o = sample_json['lat']
    p = sample_json['long']
    q = sample_json['sqft_living15']
    r = sample_json['sqft_lot15']
    s = sample_json['month']
    t = sample_json['year']
    
    columns = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade',
              'sqft_above','sqft_basement','yr_built','yr_renovated','lat','long','sqft_living15','sqft_lot15','month',
              'year']
    
    house = [[b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t]]
    
    house = pd.DataFrame(house, columns=columns)
    
    house = scaler.transform(house.values.reshape(-1, 19))
    
    prediction = model.predict(house)
    
    return float(prediction)
```

## I use the function for prediction

```python
house = df.drop('price',axis=1).iloc[0]
house
```

```python
(return_prediction(model,scaler,house))
```
